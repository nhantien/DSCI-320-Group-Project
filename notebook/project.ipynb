### **Merged Datasets**

1. [Amazon bestsellers](https://www.kaggle.com/datasets/sootersaalu/amazon-top-50-bestselling-books-2009-2019) 
   - "Dataset on Amazon's Top 50 bestselling books from 2009 to 2019. Contains 550 books, data has been categorized into fiction and non-fiction using Goodreads."  
2. [Conlit](https://figshare.com/articles/dataset/CONLIT/21166171/1?file=37535605)
   - "This dataset includes derived data on a collection of ca. 2,700 books in English published between 2001-2021 and spanning twelve different genres."
3. [Goodreads best books ever](https://zenodo.org/record/4265096#.ZAgSxOzMKvA)
   - "The dataset contains 25 variables and 52478 records corresponding to books on the GoodReads Best Books Ever list (the larges list on the site)."
4. [New York Times (NYT) bestsellers](https://www.kaggle.com/datasets/dhruvildave/new-york-times-best-sellers)
   - "The data contains Best Sellers List published by The New York Times every Sunday. The temporal range is from 03-Jan-2010 to 29-Dec-2019 which makes it a whole decade of data. Each week, 5 books are named as best sellers for each category."

```python
import numpy as np
import pandas as pd
import altair as alt
from altair import datum
import ast
from datetime import datetime


# Handle large data sets without embedding them in the notebook
alt.data_transformers.enable('data_server')
# # Include an image for each plot since Gradescope only supports displaying plots as images
# alt.renderers.enable('mimetype')
```

```python
url = "https://github.com/kemiolamudzengi/dsci-320-datasets/blob/main/amazon_conlit_goodreads_nyt.csv?raw=true"
books = pd.read_csv(url, parse_dates=['amazon_year', 'conlit_pubdate', 'nyt_published_date',
                                      'goodreads_publish_date', 'goodreads_first_publish_date'
                                     ] )
```

Change the data type in `goodreads_genres` and `goodreads_awards` from string representation of list into actual list.

For example: `"[1, 2, 3]"` -> `[1, 2, 3]`

```python
books['goodreads_genres'] = books.goodreads_genres.apply(lambda x: ast.literal_eval(x))
books['goodreads_awards'] = books.goodreads_awards.apply(lambda x: ast.literal_eval(x))
```

### Correlation matrix of numerical attributes

```python
cor_books = (books.corr().stack().reset_index().rename(columns ={0:'correlation', 'level_0': 'var1','level_1': 'var2'}))
cor_books['correlation_label'] = cor_books['correlation'].map('{:.2f}'.format)
cor_plot = alt.Chart(cor_books).mark_rect().encode(
    alt.X('var1'),
    alt.Y('var2'),
    alt.Color('correlation_label:Q')
).properties(
    height = 500,
    width=500,
    title ='Correlation matrix of numerical attributes'
)
num = cor_plot.mark_text().encode(
    text='correlation_label',
    color=alt.condition(
        alt.datum.correlation > 0.5, 
        alt.value('white'),
        alt.value('black'))
)
cor_plot+num
```

#### Comment: 

We can see from the correlation matrix that pairs (conlit_total_ratings, amazon_num_reviews), (goodreads_likedPercent, goodreads_rating) and (amazon_rating, goodreads_rating) have high positive correlations, which is reasonable and to our expectation as books that have ratings on one website should also have good ratings on another websites, same applies for number of reviews/ratings. Other than that, pairs (goodreads_num_ratings, amazon_num_reviews) and (amazon_rating, goodreads_likedPercent) also have relatively good positive correlations. However, to our surprise, goodreads_price and amazon_price have very low correlation, which we will further investigating with one of our tasks. Moreover, we also see that nyt_weeks_on_list does not have high correlation with any of the variables in the plot above, which suggests books with more ratings/likes/reviews do not necessary stay on best-sellers list longer. We will further looking into nyt_weeks_on_list with respect to number of awards in our task.

### Most popular book genres over the years?

```python
books['conlit_pubyear'] = books['conlit_pubdate'].dt.year

dropdown_options = sorted(
    books['conlit_pubyear'].dropna().unique(), 
    reverse=True
)

year_selector = alt.binding_select(
    options=dropdown_options,
    name = 'Year: '
)
year_dropdown = alt.selection_single(
    fields=['conlit_pubyear'], 
    bind=year_selector
)
## line chart over time

genre_line = alt.Chart(books.dropna(subset=['conlit_genre'])).mark_line().encode(
    alt.X('year(conlit_pubdate):O'),
    alt.Y('count()'),
    alt.Color('conlit_genre')
).properties(
    width = 600,
    height = 250
)
# bar chart
genre_bar = alt.Chart(books.dropna(subset=['conlit_genre'])).transform_filter(
    year_dropdown
).mark_bar().encode(
    alt.X('count()', axis=alt.Axis(tickMinStep=1)),
    alt.Y('conlit_genre', sort='-x')
).add_selection(
    year_dropdown
).properties(
    width=600,
    height=250
)
```
